```python
import pandas as pd

df_orig = pd.read_csv("raeresek_eredeti_igazsagossag.csv",encoding="utf-8")

df_orig.head()
```

```python
n_shifts = df_orig.iloc[1,:].nunique()
n_workers = int((df_orig.shape[1]-1)/n_shifts)
n_days = df_orig.shape[0]-3
print([n_shifts, n_workers,n_days])
```

```python
df = pd.DataFrame(columns=["Datum","Muszak"] + [df_orig.columns[i * 4 + 1] for i in range(n_workers)])

for i in range(n_days):
    for j in range(n_shifts):
        df.loc[df.shape[0],'Datum'] = df_orig.iloc[i+1,0]
        df.loc[df.shape[0]-1,'Muszak'] = df_orig.iloc[0,j+1]
        for k in range(n_workers):
            df.loc[df.shape[0]-1,df.columns[k+2]] = df_orig.iloc[i+1,k*n_shifts+j+1]

df.head(5)
```

```python
df.loc[df.shape[0],'Datum'] = "Min_ora"
df.loc[df.shape[0],'Datum'] = "Max_ora"
def_val_1 = 56
def_val_2 = 240
for k in range(n_workers):
    value_1 = df_orig.iloc[df_orig.shape[0]-2,k*n_shifts+1] 
    value_2 = df_orig.iloc[df_orig.shape[0]-1,k*n_shifts+1]
    df.loc[df.shape[0]-2,df.columns[k+2]] = value_1 if not pd.isna(value_1) else def_val_1
    df.loc[df.shape[0]-1,df.columns[k+2]] = value_2 if not pd.isna(value_2) else def_val_2

df.iloc[[df.shape[0]-2,df.shape[0]-1],:]
```

```python
df.replace({'I': 0, 'N': 1000, 'T':1}, inplace=True)
df.to_csv('raeresek.csv', index=False, encoding='utf-8')
```
